***Bollinger Bands Calculations***

```python
tickerSymbol = 'TATAMOTORS.NS'


tickerData = yf.Ticker(tickerSymbol)


df = tickerData.history(period='1h', start='2022-1-1', end='2023-12-31')


print(df)

df['20 Day MA'] = df['Close'].rolling(window=20).mean()
df['20 Day STD'] = df['Close'].rolling(window=20).std()
df['Upper Band'] = df['20 Day MA'] + (df['20 Day STD'] * 2.5)
df['Lower Band'] = df['20 Day MA'] - (df['20 Day STD'] * 2.5)

plt.figure(figsize=(10, 5))
#plt.plot(df['Close'], color='blue')
plt.plot(df['20 Day MA'],color='red')
plt.plot(df['Upper Band'],color='green')
plt.plot(df['Lower Band'],color='grey')
```

# ***RSI & EMA Based Trading Strategy***

```python
def calculate_ema(data, window):
    ema = [data.iloc[0]]
    multiplier = 2 / (window + 1)

    for i in range(1, len(data)):
        ema_value = (data.iloc[i] - ema[i - 1]) * multiplier + ema[i - 1]
        ema.append(ema_value)

    return ema

def calculate_sma(data, window):
    sma = [data.iloc[0]]

    for i in range(1, len(data)):
        sma_value = data.iloc[i - window + 1:i + 1].mean()
        sma.append(sma_value)

    return sma

def calculate_rsi(data, window=14):
    delta = data.diff()

    gains = delta.where(delta > 0, 0)
    losses = -delta.where(delta < 0, 0)

    avg_gain = gains.rolling(window=window, min_periods=1).mean()
    avg_loss = losses.rolling(window=window, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    return rsi


df = yf.download('TATAMOTORS.NS', start='2019-01-01', end='2023-01-01')


ema_5 = 5
ema_10 = 10
ema_7 = 7
ema_15 = 15

sma_50 = 50
sma_100 = 100

rsi_buy_lower = 30
rsi_buy_upper = 55

rsi_sell_lower = 45
rsi_sell_upper = 70


df['EMA_5'] = calculate_ema(df['Close'], ema_5)
df['EMA_10'] = calculate_ema(df['Close'], ema_10)
df['EMA_7'] = calculate_ema(df['Close'], ema_7)
df['EMA_15'] = calculate_ema(df['Close'], ema_15)

df['SMA_50'] = calculate_sma(df['Close'], sma_50)
df['SMA_100'] = calculate_sma(df['Close'], sma_100)


df['RSI'] = calculate_rsi(df['Close'])

#Buy and sell signal
df['Signal'] = 0
df.loc[(df['EMA_5'] > df['EMA_15']) & (df['RSI'] >= rsi_buy_lower) & (df['RSI'] <= rsi_buy_upper) , 'Signal'] = 1
df.loc[(df['EMA_7'] < df['EMA_15']) & (df['RSI'] >= rsi_sell_lower) & (df['RSI'] <= rsi_sell_upper), 'Signal'] = -1


print(df)


plt.figure(figsize=(10, 5))
plt.plot(df['Close'], label='Close Price', color='black')
plt.scatter(df.index[df['Signal'] == 1], df['Close'][df['Signal'] == 1], marker='^', color='g', label='Buy Signal')
plt.scatter(df.index[df['Signal'] == -1], df['Close'][df['Signal'] == -1], marker='v', color='r', label='Sell Signal')
plt.title('Stock Price with Buy and Sell Signals')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
